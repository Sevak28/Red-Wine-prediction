```python
x=ds[['alcohol','sulphates','citric acid','fixed acidity','pH','residual sugar','chlorides']]
y=ds['quality']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
from sklearn.preprocessing import StandardScaler
SE = StandardScaler()
x_train,x_test, y_train, y_test = tts(x,y,random_state = 42)
SE.fit(x_train)
x_train = SE.transform(x_train)
x_test = SE.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier as kc
```

```python
n_neigh = len(np.unique(ds['quality']))
```

```python
model = kc(n_neighbors = n_neigh)
model.fit(x_train,y_train)
print(f'Average K neighbors precision {model.score(x_test,y_test)}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2 = RandomForestClassifier(max_depth =30 , n_estimators = 200,random_state= 42)
model2.fit(x_train,y_train)
print(f'Average Random Forest precision {model2.score(x_test,y_test)}')
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
prediction2 = model2.predict(x_test)
print(classification_report(y_test,prediction2))
plot_confusion_matrix(model2, x_test, y_test) 
plt.show()
```
